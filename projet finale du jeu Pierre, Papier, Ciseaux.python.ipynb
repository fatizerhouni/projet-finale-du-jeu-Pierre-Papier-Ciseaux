```python
import tkinter as tk
from random import choice as random_choice
```

```python


# Fonction pour déterminer le gagnant
def determine_winner(player_choice, computer_choice):
    if player_choice == computer_choice:
        return "Draw"
    elif (player_choice == "Pierre" and computer_choice == "Ciseaux") or \
         (player_choice == "Feuille" and computer_choice == "Pierre") or \
         (player_choice == "Ciseaux" and computer_choice == "Feuille"):
        return "You Win"
    else:
        return "Computer Wins"

# Fonction pour mettre à jour les scores et le résultat
def play(player_choice):
    computer_choice = random_choice(["Pierre", "Feuille", "Ciseaux"])
    result = determine_winner(player_choice, computer_choice)
    
    if result == "You Win":
        scores["player"] += 1
    elif result == "Computer Wins":
        scores["computer"] += 1
    
    result_label.config(text=f"Computer chose {computer_choice}. {result}!")
    your_score_label.config(text=f"YOU {scores['player']}")
    computer_score_label.config(text=f"COMPUTER {scores['computer']}")

# Initialisation de la fenêtre Tkinter
window = tk.Tk()
window.geometry("400x400")
window.title("Pierre-Feuille-Ciseaux")
root = tk.Frame(window)

# Scores
scores = {"player": 0, "computer": 0}

# Cadre et étiquette pour le score du joueur
your_score_frame = tk.Frame(root, bg='blue')
your_score_label = tk.Label(your_score_frame, text="YOU 0", font=("Arial", 14))

your_score_label.pack(padx=10, pady=10)
your_score_frame.grid(row=0, column=0, padx=10, pady=10)

# Cadre et étiquette pour le score de l'ordinateur
computer_score_frame = tk.Frame(root, bg='red')
computer_score_label = tk.Label(computer_score_frame, text="COMPUTER 0", font=("Arial", 14))

computer_score_label.pack(padx=10, pady=10)
computer_score_frame.grid(row=0, column=1, padx=10, pady=10)

# Cadre et étiquette pour afficher le résultat
result_frame = tk.Frame(root)
result_label = tk.Label(result_frame, text="", font=("Arial", 14))

result_label.pack(padx=10, pady=10)
result_frame.grid(row=1, column=0, columnspan=2)

# Cadre pour les boutons
buttons_frame = tk.Frame(root)
choices = ["Pierre", "Feuille", "Ciseaux"]

for i, player_choice in enumerate(choices):
    button = tk.Button(buttons_frame, text=player_choice, font=("Arial", 14), command=lambda player_choice=player_choice: play(player_choice))
    button.grid(row=0, column=i, padx=10, pady=10)

buttons_frame.grid(row=2, column=0, columnspan=2)

# Ajouter le cadre principal
root.pack()

# Boucle principale de Tkinter
window.mainloop()
```
